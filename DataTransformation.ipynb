# Part #1

### a) Plot the Total Energy Consumed by the Residential Sector per month.

```python
# Converting Energy consumtion to float
ec_df["Total Energy Consumed by the Residential Sector"] = ec_df["Total Energy Consumed by the Residential Sector"].astype(np.float)
```

```python
mg = ec_df[["Month","Total Energy Consumed by the Residential Sector"]].plot(kind = 'line', figsize=(10,7))
mg.set_title('Monthly Residential Energy Consumed by the Residential Sector')
mg.set_xlabel('Monthly')
mg.set_ylabel('Residential Energy Consumption');
```

### b) Aggregate the data above into yearly values, and plot the yearly residential Sector per month

```python
# Splitting Year and Month from Month column
ec_df[['Year','Month']] = ec_df.Month.str.split(" ",expand=True,)
```

```python
# dataframe has Month and Year columns now
ec_df['Year'].astype(np.int)
```

```python
# Aggregating yearly power consumption
df_yearly = ec_df.groupby(["Year"])["Total Energy Consumed by the Residential Sector"].apply(sum).reset_index()
```

```python
df_yearly
```

```python
yg = df_yearly.plot(kind = 'line', figsize=(10,7))
yg.set_title('Yearly Residential Energy Consumed by the Residential Sector')
yg.set_xlabel('Yearly')
yg.set_ylabel('Residential Energy Consumption');
yg.set_xlim(0,50)
yg.set_ylim(14000,22000)
```

### a) Load the wine dataset using the instructions here: https://scikit-learn.org/stable/datasets/index.html

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import preprocessing
from sklearn.decomposition import PCA
```

```python
wine = datasets.load_wine()
```

```python
wine_df = pd.DataFrame(data= np.c_[wine['data'], wine['target']],
                     columns= wine['feature_names'] + ['target'])
```

```python
wine_df
```

```python
wine_df.info()
```

```python
# Separating out the features and target before applying scalings and PCA
X_df = wine_df.iloc[:,:-1] # features
y_df = wine_df.iloc[:,-1:] # target
```

```python
# Will apply scalling and PCA on this features dataframe
X_df
```

```python
X_df.describe()
```

### b) Apply the MinMaxScaler() from the sklearn preprocessing library to the wine data to normalize it. 

```python
# Applying minmaxScaler()
wine_df_cp = X_df
mm_scaler = preprocessing.MinMaxScaler()
wine_mm = mm_scaler.fit_transform(wine_df_cp)
wine_mm # converted to Numpy array
```

```python
# Convert Numpy array in dataframe for comparison before and after minmaxScaler
wine_df_mm = pd.DataFrame(data=wine_mm[0:,0:],    # values
             #index=wine_df_mm[1:,0],    # 1st column as index
             columns=X_df.columns)
```

```python
# Can see min and max values of all feature in the range of 0 to 1
# Now all features are all on the same relative scale
wine_df_mm.describe()
```

```python
wine_df_mm
```

### c) Apply the StandardScaler() from the sklearn preprocessing library to the wine data to standardize it

```python
# Applying StandardScaler()
wine_df_cp2 = X_df
ss_scaler = preprocessing.StandardScaler()
wine_ss = ss_scaler.fit_transform(wine_df_cp2)
wine_ss # converted to Numpy array
```

```python
# Convert Numpy array in dataframe for comparison before and after StandardScaler
wine_df_ss = pd.DataFrame(data=wine_ss[0:,0:],    # values
             #index=wine_df_mm[1:,0],    # 1st column as index
             columns=X_df.columns)
```

```python
# Can see min and max values of all feature in the range of -2 to 4
# Also std dev is 1 for all features
wine_df_ss.describe()
```

```python
wine_df_ss
```

### d) Apply the PCA library to the standardized data from c) and determine the number of dimensions that can be reduced while capturing 90% of the variance in the data.   State the answer to this question explicitly in a comment in the Jupyter notebook following the computation.

```python
# PCA for standardized data with all components = 13
pca = PCA()
```

```python
# Applying fit transform on standardized data
wine_pca_ss = pca.fit_transform(wine_df_ss)
```

```python
# Getting cumulative sum of features 1 to 13. Value of index 0 shows variance if only 1 dimension is selected.
# Similarly, each index (n) of the array shows variance with the (n+1) dimensions selected.
pca.explained_variance_ratio_.cumsum()
```

### As shown above, if we select n = 7 (92%), we can get atleast 90% variance. Therefore, we need to select atleast 8 dimensions. Selecting 8 (n+1) columns for PCA

```python
# PCA for minmax data with all components = 13
# Applying fit transform on minmax data
wine_pca_mm = pca.fit_transform(wine_df_mm)
```

```python
pca.explained_variance_ratio_.cumsum()
```

### As shown above, minmax data also gives the same output for selection of number of dimensions = 8.

```python
# Trying PCA with 8 dimensions for minmax data
pca=PCA(n_components=8)
wine_pca_mm=pca.fit_transform(wine_df_mm)
wine_df_pca_mm=pd.DataFrame(data = wine_pca_mm[0:,0:], columns = wine_df_mm.columns[0:8])
wine_df_pca_mm
```

```python
# Trying PCA with 8 dimensions for standardized data
pca=PCA(n_components=8)
wine_pca_ss=pca.fit_transform(wine_df_ss)
wine_df_pca_ss=pd.DataFrame(data = wine_pca_ss[0:,0:], columns = wine_df_ss.columns[0:8])
wine_df_pca_ss
```
