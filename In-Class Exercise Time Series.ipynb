```python
import matplotlib.pyplot as plt
import pandas as pd
import quandl

quandl.ApiConfig.api_key = '***' # as soon as you sign up on https://www.quandl.com, you will get an API key 

# get stock data

data1 = quandl.get('WIKI/****') # use your favorite stock symbol, e.g., AAPL

data1.head(5)
```

```python
y = data1['Open']
y.plot(figsize=(15, 6))
plt.show()
```

```python
diffed_y = pd.DataFrame(y.diff().dropna())

type(diffed_y)
diffed_y.plot(figsize=(15, 6))
plt.show()
```

```python
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_pacf
#series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0)
plot_pacf(y, lags=50)
pyplot.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(y, lags = 50)
pyplot.show()
```

```python
plot_acf(diffed_y, lags = 25)
```
